# **K Nearest Neighbors**



Iris dataset [here](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html#sklearn.datasets.load_iris) \
```k-NN``` documentation [here](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html#sklearn.neighbors.KNeighborsClassifier)

````
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier as kNN
from sklearn.datasets import load_iris # Iris dataset
````

````
iris = load_iris() # Load the iris dataset, NOTA ver la informacion del return de la documentacion de iris
iris_x = iris.data[:, :2] # Features. We'll pick 2 NOTA los : significa toda la fila y :2 significa 2 columas con el fin de seleccionar ddos dimensopmes
iris_y = iris.target # Labels
features_names = iris.feature_names

# Print shapes
print (iris_x.shape, iris_y.shape, features_names)
_, ax = plt.subplots(figsize=(6,5))
ax.scatter (iris_x[:,0], iris_x[:,1], c = iris_y, cmap = "tab20")
````

````
# Instance kNN object
n_neighbors = 5
knn = kNN(n_neighbors = n_neighbors , metric = "euclidean")

# Training
knn.fit( iris_x , iris_y ) # (num_samples, num_features), (num_samples)
````

````
## Plot decision boundaries

#hacer una funcion que se encargue de graficar las fronteras de decicion

h_min, h_max = iris_x[:,0].min()-0.1, iris_x[:,0].max()+0.1
v_min, v_max = iris_x[:,1].min()-0.1, iris_x[:,1].max()+0.1

h_grid, v_grid = np.meshgrid(np.linspace(h_min, h_max, 100), np.linspace(v_min, v_max, 100))

#print (h_grid.shape, v_grid.shape, h_grid.reshape(-1).shape, v_grid.reshape(-1).shape)

#print (np.c_[h_grid.reshape(-1), v_grid.reshape(-1)].shape)

pred_grid = knn.predict(np.c_[h_grid.reshape(-1), v_grid.reshape(-1)])
print (pred_grid.shape)
pred_grid = pred_grid.reshape(h_grid.shape)
print(pred_grid.shape, h_grid.shape, v_grid.shape)
````

````
_, ax = plt.subplots (figsize=(6,5))
ax.pcolormesh(h_grid, v_grid, pred_grid, cmap = "Dark2")
ax.scatter(iris_x[:, 0], iris_x[:, 1], c=iris_y, cmap = "Dark2", edgecolor="k")

ax.set_xlabel(features_names[0])
ax.set_ylabel(features_names[1])
````
